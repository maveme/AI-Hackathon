## <u>STEP 3: DEFINE YOUR PROBLEM</u>
This step is a more theoretical yet crucial part of this tutorial. Take it as a little introduction to the modeling universe. \
First, you have to define your problem. That is, what do you want the agent to do? 
Once defined the actions that are going to characterize your bot, think of the input and the output these actions will have. You have to consider what triggers the bot such that an action will happen. Specifically, you are going to be the one creating and manipulating those triggers. 

I will now give you an example based on PIZZA_BOT. 
- **What do we want our bot to do?** We want our PIZZA_BOT to take pizza orders from customers calling a pizzeria. 
- **What kind of interaction do we expect?** Our bot will answer the phone like a pizzeria employee. It will ask for customer's name and take the name of the pizza she/he wants to order. It will ask if waiting time is ok and confirm order if customer agrees. The bot will ask for delivery address after order confirmation. 
- **Which data we want to collect?** The main information we want to collect is user name, user address, pizza name and pizza size. 

This step is crucial because it gives you a general overview of the human-bot interaction you want to develop. 
Let's go on with the next step.

## <u>STEP 4: INTENTS</u>
First of all, congratulations on making it this far. You're doing an amazing job! The next step is about creating intents for our agent. 


<a href="https://cloud.google.com/dialogflow/es/docs/intents-overview?hl=en">Short definition:</a>
"An intent categorizes an end-user's intention for one conversation turn. For each agent, you define many intents, where your combined intents can handle a complete conversation. When an end-user writes or says something, referred to as an end-user expression, Dialogflow matches the end-user expression to the best intent in your agent. Matching an intent is also known as intent classification."

For our project, we need different intents:
- a customized **welcome intent**, that makes the bot answer the call as a pizzeria employee
- a **pizza_call intent**, that asks for customer's name and the asks the customer how it can help. 
- a **pizza intent**, that gets the information relative to the pizza the customer wants to order (pizza_name and pizza_size). It then confirms the order and tells the customer the expected waiting time and ask if that's okay for the customer. 
- a **no intent**, if the customer does not want to wait that much time. The bot then ends the conversation politely. 
- a **yes intent**, if the customer wants to wait that much time. At this point, the agent takes the information relative to customer's address. Then it ends the conversation. 


Before we start creating new intents, delete the default welcome one.
<br>
<center><img src="images/8.jpg" alt="" width="700"/></center>

The first intent we want to create is named **Welcome**. In this case we want the intent to be triggered without the need for matched text or spoken input. We can do this by using one of the default **events** on the platform, in this case the "welcome" event. 
We also want the agent to deliver a text rich **response** to the user, "Hi, thank you for calling pizzeria Italia!". 
<br>
<center><img src="images/9.jpg" alt="" width="700"/></center>

It's now time to get to the core intents of our interaction.\
The second intent it's going to be named **Pizza_Call**. This intent will be triggered by some sentences that we expect the customer to say.  The set of the triggering sentences is represented by "Training Phrases". 
<br>
<center><img src="images/10.jpg" alt="" width="700"/></center>
<br>

Now we start defining **actions** for our intents. You can define an action for each intent. When an intent is matched, Dialogflow provides the action to your system, and you can use the action to trigger certain actions defined in your system. For each action we also define a **parameter**. When an intent is matched at runtime, Dialogflow provides the extracted values from the end-user expression as parameters. Each parameter has a type, called the entity type, which dictates exactly how the data is extracted. Unlike raw end-user input, parameters are structured data that can easily be used to perform some logic or generate responses. Our parameter will be required, so that our intent will be incompleted without the corresponding parameter. In the **prompts**, we write the questions the agent will ask user if a request does not contain this parameter. In our case, the questions are going to ask for user name and store the information in the parameter user_name.   We also define the message the agent will deliver to the user.
